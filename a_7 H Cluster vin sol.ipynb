#  Question 1
   Eastwest Airlines Passengers details

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
air=pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
air
```

```python
air.drop(columns=["ID#"],axis=1,inplace=True)
```

as we do not required ID coloumn we can drop that

```python
air
```

```python
air.info()
```

```python
air.isnull().sum()
```

```python
air.shape
```

Normalising our data

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
air1 = pd.DataFrame(trans.fit_transform(air.iloc[:,0:]))
air1   
```

```python
# for creating dendrogram
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch 
z = linkage(air1, method="average",metric="euclidean")
plt.figure(figsize=(20, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
)   
plt.show()     
```

```python
from sklearn.cluster import AgglomerativeClustering 
import warnings 
warnings.filterwarnings('ignore')
h_complete = AgglomerativeClustering(n_clusters=5, linkage='average',affinity = "euclidean").fit(air) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
air['clust']=cluster_labels # creating a  new column and assigning it to new column 
air 
```

```python
air.iloc[:,1:].groupby(air.clust).mean()  
```

By the above information we can understand that the passengers of Eastwest airlines who are belong to cluster 0 have an avg of 95974.26  miles eligible for award travel and eligible for award.

Like this we can analyse for all the clusters


```python
clus0 = air[(air.clust==0)]
clus0  
```

The above data is belongs to the passengers who are in cluster 0

```python
clus1 = air[(air.clust==1)]
clus1
```

The above data is belongs to the passengers who are in cluster 1

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
 
z = linkage(air1, method="single",metric="euclidean")
plt.figure(figsize=(45, 15))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    
)
plt.show()       
```

when we apply the single linkage method our dendrogram looks like this

The above solutions are done by hierarchial clustering method

But The given data set file has many observations and hierarchial clustering method is not the best suitable thing to cluster this data so we have to apply some other concepts as hierarchial clustering limited to small data sets.
